```python
import pandas as pd
```

```python
import numpy as np
```

```python
from tqdm import tqdm
```

```python
train = pd.read_csv("../data/train_mod.csv")
```

```python
test = pd.read_csv("../data/test_mod.csv")
```

```python
train.shape
```

```python
test.shape
```

```python
train = train.sort_values(["year", "month"])
```

```python
train["temp"] = train.apply(lambda x : str(int(x["shop_id"])) + "_" + str(int(x["item_id"])), axis = 1 )
```

```python
train.head()
```

```python
test["item_price"] = 0
```

```python
test["temp"] = test.apply(lambda x : str(x["shop_id"]) + "_" + str(x["item_id"]), axis = 1 )
```

```python
test.head()
```

```python
test.index = test["temp"]
test.index = test["item_id"]
```

```python
test.head()
```

```python
test_temp_list = test["temp"].tolist()
test_temp_list = test["item_id"].tolist()
```

```python
train2 = train[train["temp"].isin(test_temp_list)]
train2 = train[train["item_id"].isin(test_temp_list)]
```

```python
train2.shape
```

```python
train2.head()
```

```python
## shop_id, item_id 모두 같은 애들 price로 집어넣음
## item_id만 같은 애들
for i in tqdm(range(len(train2))):
#     test.loc[train2["temp"].iloc[i], "item_price"] = train2["item_price"].iloc[i]
    test.loc[train2["item_id"].iloc[i], "item_price"] = train2["item_price"].iloc[i]
```

```python
test.describe()
```

```python
test[test['item_price']==0]
```

```python
test.index = test["temp"]
test_temp_list = test["temp"].tolist()
train2 = train[train["temp"].isin(test_temp_list)]
```

```python
for i in tqdm(range(len(train2))):
    test.loc[train2["temp"].iloc[i], "item_price"] = train2["item_price"].iloc[i]
```

```python
test[test['item_price']==0]
```

```python
test.head()
```

```python
test.to_csv("test_mod1.csv", index = False)
```
